Filtrar un valor en las filas de un dataframe

Este filtro devuelve un DataFrame que contiene solo las filas que cumplen con 
esta condición de igualdad.

```python
import pandas as pd

ventas = {
    'producto': ['Leche', 'Pan', 'Huevos', 'Jabón', 'Arroz'],
    'precio': [2.5, 1.0, 1.8, 3.2, 2.0],
    'cantidad': [10, 20, 15, 5, 25]
}

df = pd.DataFrame(ventas)
```

Supongamos que tenemos un DataFrame llamado "df" que contiene información sobre ciertos productos, y queremos filtrar los productos que tienen un precio superior a $50 y una cantidad en inventario mayor a 10 unidades. Podemos hacerlo de la siguiente manera:

```python
import pandas as pd

# Creamos el DataFrame

df = pd.DataFrame({
    'producto': ['A', 'B', 'C', 'D'],
    'precio': [25, 60, 45, 70],
    'inventario': [5, 15, 7, 20]
})
```

```python
# Filtramos los productos que cumplen ambas condiciones


filtro = (df['precio'] > 50) & (df['inventario'] > 10)
productos_filtrados = df[filtro]

print(productos_filtrados)
```

**Uso de Loc para filtrar**

Con el ejemplo anterior:

Uso de Loc:
loc es un método utilizado en Pandas para la selección de datos en un DataFrame.

La función principal de loc es seleccionar filas y columnas de un DataFrame basándose en etiquetas o valores de las etiquetas de índice. Es decir, con loc puedes seleccionar un subconjunto de datos del DataFrame que cumplen cierta condición o que se encuentran dentro de un rango específico de 
etiquetas.

```python
# Aplicamos el filtro con loc
productos_filtrados = df.loc[(df['precio'] > 50) & (df['inventario'] > 10)]

print(productos_filtrados)
```

**Funcion lambda para filtrar (operador and)**

Esta función lambda toma un parámetro x que es el DataFrame completo, y luego se aplican las condiciones de filtro utilizando el operador & para unir las dos condiciones
.

```python
# Aplicamos el filtro con funciones lambda
productos_filtrados = df[lambda x: (x['precio'] > 50) & (x['inventario'] > 10)]

print(productos_filtrados)
```

```python
import pandas as pd

# Creamos el DataFrame
df = pd.DataFrame({
    'producto': ['A', 'B', 'C', 'D'],
    'precio': [25, 60, 45, 70],
    'inventario': [5, 15, 7, 20]
})
```

```python
df
```

```python
# Aplicamos el filtro con funciones lambda y el operador lógico or
productos_filtrados = df[(df['precio'] > 50) | (df['inventario'] > 10)]

print(productos_filtrados)
```

**Usando Loc con operador lógico OR**

En este ejemplo, utilizamos el método loc para aplicar el filtro con las dos condiciones, utilizando el operador | para aplicar el operador lógico or. Esto nos dará como resultado un nuevo DataFrame con los productos que cumplen al menos una de las dos condiciones: tener un precio mayor a 50 o un inventario mayor a 10.

```python
# Aplicamos el filtro con loc y el operador lógico or
productos_filtrados = df.loc[(df['precio'] > 50) | (df['inventario'] > 10)]

print(productos_filtrados)
```

**Usando la función Lambda para filtrar con operador OR**

```python
# Aplicamos el filtro con una función lambda
productos_filtrados = df[lambda x: (x['precio'] > 50) | (x['inventario'] > 10)]

print(productos_filtrados)
```

**Filtros con indices de filas numericos**

**Ejercicio filtra los indices de 1 a 3 ( recordemos que los indices de dataframe empiezan en 0)**

```python
import pandas as pd

# Creamos el DataFrame
df = pd.DataFrame({
    'producto': ['A', 'B', 'C', 'D'],
    'precio': [25, 60, 45, 70],
    'inventario': [5, 15, 7, 20]
}, index=[10, 20, 30, 40])
```

Por ejemplo, para filtrar las filas con índices 1 y 3, y que cumplan la condición de que el precio sea mayor a 50 y el inventario mayor a 10, puedes utilizar el siguiente código:

En este caso, se utiliza iloc[[1, 3], :] para seleccionar las filas con índices 1 y 3, y luego se aplica el filtro con la condición de que el precio sea mayor a 50 y el inventario mayor a 10 utilizando loc.

Recordemos que los indices empiezan en cero(0)

```python
# Filtro con operador and y filtro por índices numéricos
productos_filtrados = df.iloc[[1, 3], :].loc[(df['precio'] > 50) & (df['inventario'] > 10), ['producto', 'precio']]

print(productos_filtrados)
```

**Filtrar filas segun indices:
    2 a 4 y de 6 a 9**

En este ejemplo, se crea un DataFrame con índices numéricos del 0 al 9 y luego se filtran los índices del 2 al 4 y del 6 al 9 utilizando el método iloc. Finalmente, se utiliza el método append para concatenar los dos subconjuntos de filas y obtener el DataFrame con los índices filtrados.

```python
import pandas as pd

# Creamos el DataFrame
df = pd.DataFrame({
    'producto': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'],
    'precio': [25, 60, 45, 70, 30, 55, 80, 40, 50, 65],
    'inventario': [5, 15, 7, 20, 10, 8, 12, 6, 9, 14]
})

```

**df.iloc[2:5]** filtra las filas del DataFrame df usando los índices de posición de fila. En este caso, 2:5 indica que se seleccionarán las filas con índices de posición 2, 3 y 4, pero no se incluirá la fila con índice 5. Es importante tener en cuenta que el índice de posición de fila comienza en 0. Entonces, **df.iloc[2:5]** devuelve un nuevo DataFrame que contiene las filas 2 a 4 del DataFrame original.

```python
# Filtramos los índices de 2 a 4 y de 6 a 9
indices_filtrados = df.iloc[2:5].append(df.iloc[6:10])

print(indices_filtrados)
```

**otra forma de hacerlo:**
    Este código selecciona las filas 2 a 4 y 6 a 8 y concatena los resultados en un solo DataFrame llamado df_resultado
    
    Fijate que no se incluye, los extremos 5 y 9 

```python
df_resultado = pd.concat([df.iloc[2:5, :], df.iloc[6:9, :]])
```

```python
df_resultado 
```

**Otra forma de Filtrar filas segun indices: 2 a 4 y de 6 a 9**

Utilizamos la función loc para seleccionar las filas del DataFrame basadas en los índices de las filas. 

La expresión dentro de los corchetes [] especifica los índices de las filas que queremos seleccionar.



Usamos los operadores lógicos & (and) y | (or) para combinar dos expresiones booleanas: (df.index >= 2) & (df.index <= 4) y (df.index >= 6) & (df.index <= 9), que seleccionan los índices de las filas 2-4 y 6-9, respectivamente.

```python
# Filtramos los índices de 2 a 4 y de 6 a 9
indices_filtrados = df.loc[(df.index >= 2) & (df.index <= 4) | (df.index >= 6) & (df.index <= 9)]

print(indices_filtrados)
```
