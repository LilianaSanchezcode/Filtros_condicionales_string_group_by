Tamaños de grupo

El número de filas en cada grupo de GroupBy objetos se puede obtener fácilmente usando la **función .size().**

Por ejemplo, desea ver cuántas filas diferentes hay disponibles en cada grupo de categoría de producto. 



Cuenta el número de filas en cada grupo.




Esto se puede obtener simplemente de la siguiente manera:

**Puede ver las similitudes entre ambos resultados:**



* **df.groupby("Product_Category").size()**
* **df.groupby("Product_Category").count()**





Los números son los mismos. Sin embargo, hay una diferencia significativa en la forma en que se calculan.





Según pandas, la **función agregada .count() cuenta solo los valores no nulos de cada columna,**



* **mientras size()  devuelve el número de filas disponibles en cada grupo,  independientemente de la presencia o ausencia de valores**



Por lo tanto, debe tener una sólida comprensión de la diferencia entre estas dos funciones antes de usarlas.



* Si  desea ver cuántos valores  no nulos están presentes en cada columna de cada grupo, use .count().



* De lo contrario, elija .size()

```python
df.groupby("Product_Category").size()
```

```python
df.groupby("Product_Category").count()
```

### Agregar múltiples columnas con diferentes funciones de agregado

Es una función que toma un conjunto de valores como entrada y devuelve un solo valor como resultado.







En el contexto de los dataframes de Pandas, estas funciones se aplican típicamente a columnas enteras, para generar un único valor resumen para cada columna.







Las funciones de agregación comunes en Pandas incluyen **sum, mean, min, max, count, median, std y var, entre otras.**





Estas funciones se pueden aplicar a una o más columnas de un dataframe utilizando el **método aggregate (o su forma abreviada agg)** junto con el nombre de la función deseada.

**Aplicando la función de agregación mean**

Después de **agrupar los datos por categoría de producto ,** suponga que:






    
    
    
    
* Desea ver cuál es el precio unitario promedio y la cantidad en cada categoría de producto.




Todo lo que necesita hacer es hacer referencia solo a estas columnas en el **GroupBy  objeto usando corchetes y aplicarles la función agregada .mean(),** como se muestra a continuación:
    
    
    
Puede obtener **el precio unitario promedio y la cantidad en cada grupo**

```python
#Create a groupby object
df_group = df.groupby("Product_Category")
#Select only required columns
df_columns = df_group[["UnitPrice(USD)","Quantity"]]
#Apply aggregate function
df_columns.mean()
```

**Nota:** Se puede tener el mismo resultado:




**df.groupby("Product_Category")[["UnitPrice(USD)","Quantity"]].mean()**

**Obtener el máximo, el mínimo, la suma y el promedio de Cantidad en cada categoría**

```python
df.groupby("Product_Category")[["Quantity"]].aggregate([min,
                                                        max,
                                                        sum,
                                                        'mean'])
```

```python
obtener el máximo, el mínimo, la suma y el promedio de Cantidad en cada categoría
```

**Contar la cantidad de ordenes para cada categoria de producto  (OrderID':'count)**


**Cantidad promedio de productos por orden (Quantity':'mean)**

El  **método aggregate()** se aplica al objeto agrupado creado por **groupby()** y toma como argumento el diccionario **function_dictionary,** lo que significa que se aplicarán las funciones **count y mean** a las columnas **OrderID y Quantity,** respectivamente.

El resultado final será un nuevo dataframe con una fila para cada categoría de producto y dos columnas: 


**OrderID** con la cantidad total de órdenes para cada categoría de producto y 


**Quantity** con la cantidad promedio de productos por orden para cada categoría de producto.

```python
function_dictionary = {'OrderID':'count','Quantity':'mean'}
df.groupby("Product_Category").aggregate(function_dictionary)
```
